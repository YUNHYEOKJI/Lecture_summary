<font color=blue>**시간관계 상 샘플링은 실제로 진행하지 마시오.**

```python
import pymc3 as pm

with pm.Model() as model:
    theta = pm.Beta('theta', 3, 2)
    obs = pm.Binomial('obs', 15, theta, observed = 5)

with model:
    step = pm.Metropolis()
    trace = pm.sample(5000, tune=2000, step=step, chains=2)
```

Metropolis-Hasting's Algorithm with Random Walk를 사용할 때, proposal distribution  𝑁(𝑥(𝑡),𝜎2) 의 분산  𝜎2 이 지나치게 작으면 자주 accept 하는 대신에 상태 공간에서 샘플간의 이동거리가 매우 작다. 따라서 target distribution 봉우리가 여러개가 있다고 하더라도 전체 분포를 커버하지 못하거나, 커버하는데 시간이 오래 걸린다. 따라서 수렴속도가 느리다.  
Metropolis-Hasting's Algorithm with Random Walk를 사용할 때, proposal distribution  𝑁(𝑥(𝑡),𝜎2) 의 분산  𝜎2 이 지나치게 크면 상태공간에서 샘플 간의 이동거리가 큰 대신에 적게 accept된다. 결과적으로 샘플들이 넓게 퍼져있지만 정돈되지 않은 형태(울퉁불퉁한 형태)가 나타나게 된다. 이 경우에도 수렴속도가 느리다.  
따라서 Metropolis-Hasting's Algorithm with Random Walk를 사용할 때, proposal distribution  𝑁(𝑥(𝑡),𝜎2) 의 분산  𝜎2 이 지나치게 작을 때와 지나치게 클 때 수렴속도가 느려지는 문제가 발생한다. 따라서 수렴속도를 빠르게 하려면 너무 작지도 않고 크지도 않은 적당한 분산을 설정하는게 중요하다.
