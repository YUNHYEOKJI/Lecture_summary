```python
n = int(input("n="))
k = int(input("k="))

cnt = 0
while n:
    if not n % k: # k로 나누어질때
        n //= k # n은 k로 나눈 몫
        cnt += 1
    else:
        cnt += n % k # 나머지 만큼 횟수 증가
        n -= n % k # 나머지 만큼 숫자의 감소
print("모듈이가 계산할 횟수: %d" %cnt)
```

```python
import time

year = int(input("연도를 입력하세요"))
month = int(input("월을 입력하세요"))
day = int(input("일을 입력하세요"))

yearsum = 0
for i in range(1, year):
    if i % 4 == 0:
        if i % 400 == 0:
            yearsum += 366
        elif i % 100 == 0:
            yearsum += 365
        else:
            yearsum += 366
    else:
        yearsum += 365
        
monthsum = 0
if year % 4 == 0 and year % 400 == 0 and year % 4 != 0: # 윤년
    for i in range(1,month):
        if i in (1,3,5,7,8,10,12):
            monthsum += 31
        elif i in (4,6,9,11):
            monthsum += 30
        elif i == 2:
            monthsum += 29
else:
    for i in range(1,month):
        if i in (1,3,5,7,8,10,12):
            monthsum += 31
        elif i in (4,6,9,11):
            monthsum += 30
        elif i == 2:
            monthsum += 28

daysum = yearsum + monthsum + day
dayofweek =  ""

if daysum % 7 == 0:
    dayofweek = "일"
elif daysum % 7 == 1:
    dayofweek = "월"
elif daysum % 7 == 2:
    dayofweek = "화"
elif daysum % 7 == 3:
    dayofweek = "수"
elif daysum % 7 == 4:
    dayofweek = "목"
elif daysum % 7 == 5:
    dayofweek = "금"
elif daysum % 7 == 6:
    dayofweek = "토"


print("%d년 %d월 %d일은 %s요일입니다." %(year,month,day,dayofweek))
print(daysum)
```
