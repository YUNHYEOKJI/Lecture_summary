- 행 번호와 열 번호를 컬럼으로 합친다.

```python
import numpy as np
index = np.column_stack([x_coo.row, x_coo.col]); index
```

## 순서가 섞여있는 경우 -> 텐서플로에서 받아주지를 못함

- 텐서플로의 SparseTensor로 변환

```python
x_train_sparse = tf.SparseTensor(index, x_coo.data, x_coo.shape)
```

- 행과 열 번호를 정렬

```python
x_train_sparse = tf.sparse.reorder(x_train_sparse)
```

- 모형 훈련에 사용한다

```python
model.fit(x_train_sparse, y_train, epochs=1)
```

# Early Stopping 실습

- callbacks에 EarlyStopping을 추가하면 각 에포크 끝마다 성능을 테스트하여 중단 여부를 결정한다.
