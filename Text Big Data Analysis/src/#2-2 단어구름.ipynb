# 단어 구름

### 단어 구름은 빈도가 높은 단어는 크게, 낮은 단어는 작게 시각화하는 방법이다. 
### wordcloud를 불러온다.

```python
from wordcloud import WordCloud
```

### WordCloud의 옵션은 여러개가 있지만, 중요 옵션은 다음과 같다.

#### font_path: 글꼴의 경로
#### max_words: 워드클라우드를 그릴 단어의 개수
#### background_color: 배경색 설정
#### width : 가로크기(픽셀 단위)
#### height: 세로크기(픽셀 단위)

### 배경은 흰색, 최대 100단어를 가로 400픽셀, 세로 300픽셀 크기로 단어 구름으로 그린다.

```python
wc = WordCloud(background_color='white', max_words=100, width=400, height=300) 
# 기본적으로 모든 단어 포함
```

### word_count를 to_dict 함수를 이용해 사전(dict) 형태로 변환한다.

```python
word_count.set_index('단어') # 표의 형식
```

```python
word_count.set_index('단어')['빈도'] # 표의 형식에서 빈도만 추출
```

```python
word_count.set_index('단어')['빈도'].to_dict() # 표의 형식에서 빈도만 추출하고 사전형태로 변환
```

```python
count_dic = word_count.set_index('단어')['빈도'].to_dict()

count_dic
```

```python
# 같은 코드 zip으로 짝지어서 dict 사전형태로 변환
count_dic1 = dict(zip(word_count['단어'],word_count['빈도']))
count_dic1
```

### 단어 구름을 그린다.

```python
cloud = wc.fit_words(count_dic)
```

### 노트북에서 단어구름을 확인한다.

```python
cloud.to_image()
```

### 단어 구름을 cloud.png 파일로 저장한다.
#### jpg는 자연스럽게 변화하는 사진같은 파일에 적합(그라데이션 그림자등) 
#### 그래프와 시각화 같은 자료는 png 포맷이 저장했을 때 깔끔하게 저장

```python
cloud.to_file('cloud.png')
```
