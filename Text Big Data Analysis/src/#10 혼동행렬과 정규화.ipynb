# 혼동행렬을 위한 모형 만들기

```python
!wget -c https://github.com/euphoris/datasets/raw/master/imdb.zip
```

```python
import pandas as pd
df = pd.read_csv('../data/imdb.zip')
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(df.review)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, df.sentiment.values, test_size=0.2, random_state=1235)

```

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, activation='sigmoid')) # 로지스틱회귀 형태로
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), 
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(x_train.A, y_train, epochs=200, validation_data=(x_test.A, y_test))
```

# 혼동행렬 실습

```python
prob = model.predict(x_test.A)
```

```python
import numpy as np
y_pred = np.where(prob > 0.7, 1, 0) # 0.7(threshold)을 넘으면 1, 아니면 0
# model.predict_classes(x_test.A) # threshold=0.5, 조절불가
```

```python
from sklearn.metrics import confusion_matrix
```

#### 혼동행렬

```python
confusion_matrix(y_test, y_pred) # 실제와 예측을 비교
# 행방향으로 0,1 -> 실제
# 열방향으로 0,1 -> 예측
```

```python
from sklearn.metrics import *
```

#### 정확도

```python
accuracy_score(y_test, y_pred)
```

#### 정밀도

```python
precision_score(y_test, y_pred, pos_label=1) 
# pos_label 디폴트값은 1, 0을 기준으로 하려면 0으로 변경
# 0번에서 얼마나 맞췄는지를 계산하려면 0으로 변경
```

#### 재현율

```python
recall_score(y_test, y_pred, pos_label=1)
```

#### F1

```python
f1_score(y_test, y_pred)
```

# ROC 곡선 실습

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
fpr, tpr, threshold = roc_curve(y_test, prob)
```

- fpr: False Positive Ratio = FP(위양성) / 음성 = 1 - 특이도 (실제 음성 중에 얼마나 틀렸는가?)
- tpr: True Positive Ratio = TP(진양성) / 양성 = 재현도(recall) = 민감도(sensitivity) (실제 양성 중에 얼마나 맞았는가?)

```python
plt.plot(fpr, tpr)
```

```python
roc_auc_score(y_test, prob)
```
