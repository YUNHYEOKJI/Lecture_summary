```python
import pandas as pd
from factor_analyzer import FactorAnalyzer

data = pd.read_csv('../data/Applicant.csv',sep=',', index_col=0) 
```

```python
data.head()
```

#### 1. 15개의 변수를 사용하여 4개의 요인을 포함하는 요인분석 모형을 추정하시오. (varimax rotation 사용)

```python
fa = FactorAnalyzer(n_factors=4, rotation='varimax').fit(data)
# 결과 출력
print('요인적재량 :\n',pd.DataFrame(fa.loadings_, index=data.columns))
```

```python
fa_loadings = pd.DataFrame(fa.loadings_, index = data.columns,
                          columns = ['f1','f2','f3','f4'])
```

```python
fa_loadings.sort_values(by=['f1','f2','f3','f4'],
                       axis=0,
                       ascending=False)
```

```python
print('공통성 :\n',pd.DataFrame(fa.get_communalities(), index=data.columns))
```
