48의 지원자에 대한 특징들을 요인분석

```python
import pandas as pd
from factor_analyzer import FactorAnalyzer

data = pd.read_csv('../data/Applicant.csv',sep=',', index_col=0) 
```

```python
data.head()
```

#### 1. 15개의 변수를 사용하여 4개의 요인을 포함하는 요인분석 모형을 추정하시오. (varimax rotation 사용)

```python
fa = FactorAnalyzer(n_factors=4, rotation='varimax').fit(data) # 직교행렬
```

```python
fa.loadings_ # 첫번째 열이 첫번째 factor에 대한 요인값.
# 행별로 설명변수별 factor의 값. 각각 f1,f2,f3,f4
```

```python
fa.loadings_.shape # 설명변수가 15개, factor가 4개 
```

```python
fa = FactorAnalyzer(n_factors=4, rotation='varimax').fit(data)
pd.DataFrame(fa.loadings_, columns=['f1','f2','f3','f4'],
             index=data.columns)
```

```python
fa.loadings = pd.DataFrame(fa.loadings_, index=data.columns, columns=['f1','f2','f3','f4'])
fa.loadings
```

```python
fa.loadings.sort_values(by=['f1','f2','f3','f4'],
                        axis = 0,
                        ascending = False)
```

- 첫번째 factor에 크게 영향을 미치는 요인: X5, X11, X8, X6, X12, X10, X13, X2, X14
- 두번째 factor에 크게 영향을 미치는 요인: X15, X1, X9
- 세번째 factor에 크게 영향을 미치는 요인: X7, X4
- 네번째 factor에 크게 영향을 미치는 요인: X3

- 마케팅 능력
